I pulled the data of video game sales until December 2016 to see if I could learn anything about player preferences with games across different regions.

```python
import pandas as pd
data = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
data
```

```python
#Let us only focus on the columns without NaN values.
df = data.dropna()
df
```

```python
#Part 1 - Sales across the World and Regions.
#Let us find which game has the highest global sales and local sales in North America, Europe, and Japan.
```

```python
#Firstly, let us fnd out which one has the highest global sales.
globalsales = df[df['Global_Sales'] == df['Global_Sales'].max()]
print('🌐', 'Most Global sales goes to', globalsales['Name'].values, 'with a total of', globalsales['Global_Sales'].values, '.')
#Oops, it seems like that the csv is just listed with the highest global sale in the top naturally 😥.

#Secondly, let us focus on the sales for each region.
#North America
NAsales = df[df['NA_Sales'] == df['NA_Sales'].max()]
print('🌎', 'Most North America sales goes to', NAsales['Name'].values , 'with a total of', NAsales['NA_Sales'].values, '.')
#Europe
EUsales = df[df['EU_Sales'] == df['EU_Sales'].max()]
print('🏰', 'Most Europe sales goes to', EUsales['Name'].values, 'with a total of', EUsales['EU_Sales'].values, '.')
#Japan
JPsales = df[df['JP_Sales'] == df['JP_Sales'].max()]
print('🗼', 'Most Japan sales goes to', JPsales['Name'].values, 'with a total of', JPsales['JP_Sales'].values, '.')
```

```python
#Let us find the bar plot of these global and local sales to gain more insights.
sales = pd.DataFrame( {"Sales": [82.53, 41.36, 28.96, 6.5]}, index = ["Global", "NA", "EU", "JP"] )
sales.plot.bar()
```

```python
#Part 1 Conclusion
print('Wow, it seems that Wii Sports really made its way to success around the world, congrats to the production team!', '🎊')
print('The sales rank of each region basically follow the size distribution of them.', '🗺️')
```

```python
#Part 2 - Players' Preferences for Game Ratings and Genres across Regions.
#For this part, let us focus on finding what game kinds and game ratings each region like according to several top sales and mean ranks.

#Let us first find the absolute most sold game genre and game rating for the 4 regions.
#Global
GLBkind = df.nlargest(5, 'Global_Sales')
GLBkind
```

```python
#North America
NAkind = df.nlargest(5, 'NA_Sales')
NAkind
```

```python
#Europe
EUkind = df.nlargest(5, 'EU_Sales')
EUkind
```

```python
#Japan
JPkind = df.nlargest(5, 'JP_Sales')
JPkind
```

```python
#Then let us see the average sales of each game genre and game rating

#Genre
group1 = df.groupby('Genre')

#Global
GLBmean = group1['Global_Sales'].agg('mean').reset_index()
GLBmean1 = GLBmean.nlargest(5, 'Global_Sales')
GLBmeanbar = pd.DataFrame( {"Global_Sales": GLBmean1['Global_Sales'].tolist()}, index = GLBmean1['Genre'].tolist() )
GLBmeanbar.plot.bar(ylim=(0,1.2))

#NA
NAmean = group1['NA_Sales'].agg('mean').reset_index()
NAmean1 = NAmean.nlargest(5, 'NA_Sales')
NAmeanbar = pd.DataFrame( {"NA_Sales": NAmean1['NA_Sales'].tolist()}, index = NAmean1['Genre'].tolist() )
NAmeanbar.plot.bar(ylim=(0,1.2), edgecolor = 'navy')

#EU
EUmean = group1['EU_Sales'].agg('mean').reset_index()
EUmean1 = EUmean.nlargest(5, 'EU_Sales')
EUmeanbar = pd.DataFrame( {"EU_Sales": EUmean1['EU_Sales'].tolist()}, index = EUmean1['Genre'].tolist() )
EUmeanbar.plot.bar(ylim=(0,1.2), edgecolor = 'green')

#JP
JPmean = group1['JP_Sales'].agg('mean').reset_index()
JPmean1 = JPmean.nlargest(5, 'JP_Sales')
JPmeanbar = pd.DataFrame( {"JP_Sales": JPmean1['JP_Sales'].tolist()}, index = JPmean1['Genre'].tolist() )
JPmeanbar.plot.bar(ylim=(0,1.2), edgecolor = 'red')
```

```python
#Rating
group2 = df.groupby('Rating')
from IPython.display import display

#Global
GLBmean2 = group2['Global_Sales'].agg('mean').reset_index()
GLBmean3 = GLBmean2.nlargest(6, 'Global_Sales')
GLBmeanbar = pd.DataFrame( {"Global_Sales": GLBmean3['Global_Sales'].tolist()}, index = GLBmean3['Rating'].tolist() )
GLBmeanbar.plot.bar(ylim=(0,2))

#NA
NAmean2 = group2['NA_Sales'].agg('mean').reset_index()
NAmean3 = NAmean2.nlargest(6, 'NA_Sales')
NAmeanbar = pd.DataFrame( {"NA_Sales": NAmean3['NA_Sales'].tolist()}, index = NAmean3['Rating'].tolist() )
NAmeanbar.plot.bar(ylim=(0,2), edgecolor = 'navy')

#EU
EUmean2 = group2['EU_Sales'].agg('mean').reset_index()
EUmean3 = EUmean2.nlargest(6, 'EU_Sales')
EUmeanbar = pd.DataFrame( {"EU_Sales": EUmean3['EU_Sales'].tolist()}, index = EUmean3['Rating'].tolist() )
EUmeanbar.plot.bar(ylim=(0,2), edgecolor = 'green')

#JP
JPmean2 = group2['JP_Sales'].agg('mean').reset_index()
JPmean3 = JPmean2.nlargest(6, 'JP_Sales')
JPmeanbar = pd.DataFrame( {"JP_Sales": JPmean3['JP_Sales'].tolist()}, index = JPmean3['Rating'].tolist() )
JPmeanbar.plot.bar(ylim=(0,2), edgecolor = 'red')
```

```python
#Part 3 - Games with Great Score from Critics and Players
#Finally, let us find the games with the highest 10 critic scores and 10 user (player) scores. 💯

#Critic Review Scores
CS = df.nlargest(10, 'Critic_Score')
CSbar = pd.DataFrame( {"Critic_Score": CS['Global_Sales'].tolist()}, index = CS['Critic_Score'].tolist() )
CSbar.plot.bar(edgecolor = 'black')
```

```python
#Player (user) Review Scores
#I need to first deal with the 'tbd (to be discussed)' value in the User_Score coumn.
df1 = df.drop( df[(df['User_Score'] == 'tbd')].index )
df1['User_Score'].convert_dtypes(infer_objects=False).dtypes
PS = df1.nlargest(10, 'User_Score')
PSbar = pd.DataFrame( {"Player_Score": PS['Global_Sales'].tolist()}, index = PS['User_Score'].tolist() )
PSbar.plot.bar(edgecolor = 'grey')
```
